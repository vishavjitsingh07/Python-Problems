```python
#Basic Operation Required for program



class Node: #Creating a Node Class
    def __init__(self, data):
        self.data = data
        self.next = None
        
        
        
class Node2: #creating a Doubly Linked List
    def __init__(self, data):
        self.data = data 
        self.next = None #Next
        self.prev = None #Previous
        
def createDoubleLL(arr):
    if len(arr) <= 1:
        return None, None
    head = None #Intiallize head with None
    
    
    temp = None #Temporary Node
    for i in range(len(arr)):
        if arr[i] == -1: #-1 Means end of Node
            break
            
        NewNode = Node2(arr[i]) #Creating a new Node
        if head is None: #Assigning Head
            head = NewNode
            temp = head
        else:
            temp.next = NewNode #Next Node Connection
            NewNode.prev = temp #Previous Node Connection
            temp = temp.next #Moving temporary pivot
        tail = temp
    return head, tail #Return Head and Tail


def createLL(a): #Creating a Linked List
    head = None #Intiallize head with None
    for i in a: #Iterate over array
        if i == -1: #-1 means end 
            break
                
        currNode = Node(i) #Curr node
        if head == None:  #Checking is None or not
            head = currNode 
            tail = currNode
        else:
            tail.next = currNode #Adding other Nodes as tail
            tail = currNode
    return head #returing the head

def printLL(head):
    if head is None: #Edge Case
        return
    while head is not None:
        print(head.data, '--->', end = " ") #Printing Node
        head = head.next
    print(None) #Just for fun no need
    
def lenLL(head): #Finding Lenght of LinkedList Iterative
    count = 0
    while head is not None: #Tail
        head = head.next
        count +=1 #iterator
    return count #Return Count

def ithNode(head, ith):    
    if ith <0: #handling -ve case
        print('i cannot be negative')
        return
    
    count = 0
    while count <= ith-1: #
        if head is None: #Handling i>len(LinkedList)
            print('i is greater than lenght of LinkedList')
            return
        head = head.next #head.next
        count += 1 #counter
        
    return head.data

def insertAtithRec(head, i, ele):
    if i == 0: #If ith position found
        newNode = Node(ele) #NewNode
        newNode.next = head #Making new Node as head
        return newNode
    
    if i<0: #if i is less than 0
        return head
    if head is None: #if i is greater than len of list
        return None
    smallHead = insertAtithRec(head.next, i-1, ele) #Getting Small Head
    head.next = smallHead #Joining New head(small head) with value before that if any
    return head


def delNode(head, i): #Delete Ith node Iteratively
    if i<0 or i>lenLL(head): #Handling Edge Cases
        return head
    if i == 0: #If 0 changing head
        head = head.next
        return head
    
    count = 0
    currNode = head
    while count<i-1 and currNode.next is not None: #Iterate till ith value
        currNode = currNode.next
        count+=1
        
    currNode.next = currNode.next.next #Deleting node
    return head 

def delnextNode(head):
    if head.next is not None: #if next node exist
        return head.next #Getting next node
    return None
```

# Reverse a LinkedList Recursively

```python
#Time complexity O(n2)
def reverseLLRec(head):
    if head is None:  #Edge case no element in list
        return
    
    if head.next is None: #base case
        return head
    
    smallhead = reverseLLRec(head.next) #traversing the list
    
    #Getting the head and assign it as None
    curr = smallhead
    while curr.next is not None:
        curr = curr.next
    curr.next = head #tail.next is Head
    head.next = None #head is None
    return smallhead #Return tail of old list which is new head of reverse List

arr = [1, 2, 3, 4, 5, -1]
linkedList = createLL(arr)
printLL(linkedList)
newList = reverseLLRec(linkedList)
printLL(newList)
```

```python
#Time complexity O(n)
def reverseLLRec2(head):
    if head is None:  #Edge case no element in list
        return
    
    if head.next is None: #base case
        return head, head #Returning head as well as tail for last element both will same
    
    smallhead, smallTail = reverseLLRec2(head.next) #traversing the list
    smallTail.next = head #tail.next is Head
    head.next = None #head is None
    return smallhead, head #Return tail of old list which is new head of reverse List


arr = [1, 2, 3, 4, 5, -1]
linkedList = createLL(arr)
printLL(linkedList)
newListhead, newListtail = reverseLLRec2(linkedList)
printLL(newListhead)
```

```python
#Time complexity O(n) and tail is given that it is head
def reverseLLRec3(head):
    if head is None:  #Edge case no element in list
        return
    
    if head.next is None: #base case
        return head #Returning head as well as tail for last element both will same
    
    smallhead = reverseLLRec3(head.next) #traversing the list
    tail = head.next #tail is Head.next means head becomes tail in new list
    tail.next = head #make head as tail
    head.next = None #head is None
    return smallhead #Return tail of old list which is new head of reverse List

arr = [1, 2, 3, 4, 5, -1]
linkedList = createLL(arr)
printLL(linkedList)
newList = reverseLLRec3(linkedList)
printLL(newList)


from matplotlib import pyplot as plt
img = plt.imread(r"C:\Users\visha\Daily Practice\Static\linkedList.jpg")
fig, ax = plt.subplots(figsize = (8, 8))
ax.imshow(img)
plt.show()
```
