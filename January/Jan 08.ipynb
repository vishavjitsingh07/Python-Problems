# ComplexNumber Operations
Perform operations like sum, multiple, conjugate and division on complex numbers

```python
class ComplexNumber:
    def __init__(self, real, imaginary):
        self.real = real
        self.imaginary = imaginary

    def returnComplexNumber(self):
        if self.imaginary >= 0:
            result = str(str(self.real) + " " + '+' + ' '+ str(self.imaginary) + 'i')
        else: 
            result = str(str(self.real) + ' '+ str(self.imaginary) + 'i')
        return result 
    
    
    def sum(self, c2):
        real_value = c2.real + self.real
        imaginary_value = c2.imaginary + self.imaginary
        return (real_value, imaginary_value)
    def printSum(self, c2):
        real, imaginary = self.sum(c2)
        c = ComplexNumber(real, imaginary)
        return c.returnComplexNumber()

    
    def sub(self, c2):
        real_value = c2.real - self.real
        imaginary_value = c2.imaginary - self.imaginary
        return (real_value, imaginary_value)
    def printSub(self, c2):
        real, imaginary = self.sub(c2)
        c = ComplexNumber(real, imaginary)
        return c.returnComplexNumber()
    
    
    def multiple(self, c2):
        real_value = (self.real * c2.real) - (self.imaginary * c2.imaginary)
        imaginary_value = (self.real * c2.imaginary) + (c2.real * self.imaginary)
        return (real_value, imaginary_value)
    def printMultiple(self, c2):
        real, imaginary = self.multiple(c2)
        c = ComplexNumber(real, imaginary)
        return c.returnComplexNumber()
    
    
    def conjugate(self):
        imaginary = -self.imaginary
        real = self.real
        return (real,imaginary)
    def printConjugate(self):
        real, imaginary = self.conjugate()
        c = ComplexNumber(real, imaginary)
        return c.returnComplexNumber()
    
    
    def division(self, c2):
        conjugateC2_real, conjugateC2_imaginary = c2.conjugate()
        conjugateC2 = ComplexNumber(conjugateC2_real, conjugateC2_imaginary)
        real_num, imginary_num = self.multiple(conjugateC2)
        real_den, imginary_den = c2.multiple(conjugateC2)
        if real_den == 0:
            print('Error')
            return (real_den, imginary_den)
        
        real = real_num/real_den
        imaginary = imginary_num/real_den
        c = ComplexNumber(real, imaginary)
        return (c.real, c.imaginary)
    def printDivision(self, c2):
        real, imaginary = self.division(c2)
        c = ComplexNumber(real, imaginary)
        return c.returnComplexNumber()

frac1 = ComplexNumber(1, 2)
frac2 = ComplexNumber(3, 4)
print('Return Complex Number' ,frac1.returnComplexNumber())
print('Conjugate', frac1.printConjugate())
print('Return Sum', frac1.printSum(frac2))
print('Return Subtraction', frac1.printSub(frac2))
print('Return Mutliple', frac1.printMultiple(frac2))
print('Return Division', frac1.printDivision(frac2))
```

# Fraction
Find the sum, division and multiplication of two fraction numbers

```python
class Fraction:
    def __init__(self, num, dem):
        self.num = num
        self.dem = dem
        if self.dem == 0:
            print("Cann't divide by 0")
            self.dem = 1
            return
        
    def printFrac(self):
        curr = min((abs(self.num)), (abs(self.dem)))
        if curr > 1:
            while curr >= 1:
                if self.num%curr == 0 and self.dem%curr == 0:
                    break
                curr -= 1
        num = int(self.num//curr)
        dem = int(self.dem//curr)
        if dem != 1 or dem !=0:
            return str(str(num) + ' / ' +  str(dem))
        elif dem == 1:
            return str(str(num))
        else:
            print('Cannot divide by 1')
        
    def sum(self,frac2):
        num = self.num*frac2.dem + self.dem*frac2.num
        dem = self.dem*frac2.dem
        return (num, dem)
    def printSum(self, frac2):
        num, dem = self.sum(frac2)
        frac = Fraction(num, dem)
        return frac.printFrac()
    
    
    def sub(self,frac2):
        num = frac2.num*self.dem - frac2.dem*self.num
        dem = self.dem*frac2.dem
        return (num, dem)
    def printSub(self, frac2):
        num, dem = self.sub(frac2)
        frac = Fraction(num, dem)
        return frac.printFrac()
    
    
    def multiple(self, frac2):
        num = self.num*frac2.num
        dem = self.dem*frac2.dem
        return (num, dem)
    def printMultiple(self, frac2):
        num, dem = self.multiple(frac2)
        frac = Fraction(num, dem)
        return frac.printFrac()

    
    def division(self, frac2):
        num = self.num*frac2.dem
        dem = self.dem*frac2.num
        return (num, dem)
    def printDivision(self, frac2):
        num, dem = self.division(frac2)
        frac = Fraction(num, dem)
        return frac.printFrac()
    
frac1 = Fraction(2 , 3)
frac2 = Fraction(1, 2)
print('Fraction:', frac1.printFrac())
print("Sum of two fraction", frac1.printSum(frac2))
print("Subtraction of two fraction", frac1.printSub(frac2))
print("Multiple of two fraction", frac1.printMultiple(frac2))
print("Division of two fraction", frac1.printDivision(frac2))
```
