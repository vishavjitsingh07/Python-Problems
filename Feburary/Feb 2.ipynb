```python
#Helper function
import sys #Depedencies
import queue 
import random

def heightOfTree(root):
    if root is None: #base Case =  we get a leaf
        return 0 
    
    left = heightOfTree(root.left) #Left Height
    right = heightOfTree(root.right) #Right Height
    return 1 + max(left, right) #Maximum Height of either side + 1 for root


def printTree(root):
        height = heightOfTree(root) #Know the height of tree
        q = queue.Queue()
        if root is None: #if tree is Empty
            print('root is None')
            return None
        
        value = root
        q.put(value)#Add root to queue
        
        print((2**height - 1) *" " + str(root.data)) #Left distance will be 2^h - 1
        layer = 1
        count = 0
        height = height-1 #Decrement Height as goes down
        print((2**height - 1) *" ", end = '') #Distance for left child is 2^h - 1 as it is in 2nd level
        
        
        while not q.empty():
            currNode = q.get() #Get the curr Node
            leftChild = currNode.left #get the left of current Node
            if leftChild != None: #if Left is Not Node
                print(leftChild.data, end = "") #Print child
                q.put(leftChild)
            else: #else left space
                print("-", end = "")
            print((2**(height+1)-1) *" ", end = '')
            
            #if right child found print it else left space for it
            rightChild = currNode.right 
            if rightChild != None:
                print(rightChild.data, end = "")
                q.put(rightChild)
            else:
                print("-", end = "")
                
                
            print((2**(height+1) - 1) *" ", end = '')  #Distance between right child and siblings left child will be same as upper level
            
            count += 2 #Increament by 2 as left and right both covered
            if count == 2 ** layer: #if a level is completed
                layer+=1 #Increse level or layer
                height -= 1 #Decrese depth
                count = 0 #Reset counter
                print(int(((2**height)-1)) *" ", end = '') #Set right gap
                print() #\n for next layer
                print(int(((2**height)-1)) *" ", end = '') #Set left gap
                
                
            
            
            
def printTreeLevel(root):
        q = queue.Queue() #Creating a Queue
        if root is None: #if tree is empty
            print('root is None')
            return None
        
        value = root 
        q.put(value) #Add root to queue
        
        while not q.empty():#Loop till element exist in queue
            currNode = q.get()#Current Node
            leftChild = currNode.left#Getting the left child
            print(currNode.data, end = " == ") #Parent of Node
            if leftChild != None:# or leftChild.data != -1:
                print('L:', leftChild.data, end = " ") #Print the left child
                q.put(leftChild) #Add it to queue
                
            rightChild = currNode.right #getting the right child
            if rightChild != None:# or rightChild.data != -1:
                print('R:', rightChild.data, end = " ")#Print the right child
                q.put(rightChild)#Add right child to queue
            print()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        self.nodeNumber = 0
        
class BST(Node):
    def __init__(self):
        self.root = None
        self.numberOfNodes = 0
        
        
        
    
    def printtree(self):
        return printTree(self.root)
    def printtreeLevel(self):
        return printTreeLevel(self.root)
    
    
    
        
    def insertNode(self, root, node, lastCount): #Function to add a Node in tree
        if root.left is None and root.right is None: #if there is no child
            if node.data < root.data: #Left Child
                root.left = node 
                root.left.nodeNumber = lastCount #Index of Node
                lastCount += 1 
            else: #Right Child
                root.right = node
                root.right.nodeNumber = lastCount
                lastCount += 1

        elif root.left is None and root.data > node.data:
            root.left = node 
            root.right.nodeNumber = lastCount
            lastCount += 1

        elif root.right is None and root.data <= node.data:
            root.right = node
            root.right.nodeNumber = lastCount
            lastCount += 1

        else:
            if root.data > node.data:
                self.insertNode(root.left, node, lastCount+1)
            else:
                self.insertNode(root.right, node, lastCount+1)
        return root

    def createTree(self, arr):
        if len(arr) == 0:
            return None
        
        self.root = Node(arr[0])
        lastCount = 1
        self.root.nodeNumber = lastCount

        for i in arr[1:]:
            self.insertNode(self.root, Node(i), lastCount)
        return self.root
    
           
        
    
    
    def nodeSearchHelper(self, root, value):
        if root is None: #Base case 
            return False

        if root.data == value: #If node found
            return True
        elif value >= root.data and root.right is not None: #Check on right subtree
            return self.nodeSearchHelper(root.right, value)
        elif value <= root.data and root.left is not None:  #Check on left subtree
            return self.nodeSearchHelper(root.left, value) 

        return False

    def nodeSearch(self, value):
        return self.nodeSearchHelper(self.root, value)
    
    
    
    
    
    def nodePathHelper(self, root, value):
        if root is None:
            return None

        if root.data == value:
            arr = []
            arr.append(root.data)
            return arr
        elif value >= root.data and root.right is not None: #Check on right subtree
            arr = self.nodePathHelper(root.right, value)
            if arr is not None:
                arr.append(root.data)
            return arr
        elif value <= root.data and root.left is not None:  #Check on left subtree
            arr = self.nodePathHelper(root.left, value)
            if arr is not None:
                arr.append(root.data)
            return arr
    
    def nodePath(self, value):
        return self.nodePathHelper(self.root, value)   
    
    
    
    
    
    def minimumData(self, root):
        if root is None:
            return sys.maxsize
        if root.left is None:
            return root.data
        return self.minimumData(root.left)
        
    def deleteionHelper(self, root, data):
        if root is None:
            return False, None
        
        if root.data <  data:
            deleted,  newRight = self.deleteionHelper(root.right, data)
            root.right = newRight
            return deleted, root
        if root.data> data:
            deleted, newLeft = self.deleteionHelper(root.left, data)
            root.left = newLeft
            return deleted, root
        
        if root.left == None and root.right == None:
            return True, None
        
        if root.left == None:
            return True, root.right
        
        if root.right is None:
            return True, root.left
        
        
        replacement = self.minimumData(root.right)
        root.data = replacement
        
        deleted, newRoot = self.deleteionHelper(root.right, replacement)
        root.right = newRoot
        return True, root
        

    def deletion(self, value):
        self.deleteionHelper(self.root, value)
        print("Node {} is deleted".format(value))
        return self.root
```
