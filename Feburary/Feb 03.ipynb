```python
class Node: #Create a class of node having data and adress of children
    def __init__(self, data):
        self.data = data #Data of node 
        self.children = [] #List containing all the child node address
        
class Tree(Node): #tree Class having basic functions
    def __init__(self):
        self.root = None #Intialzing root with None
        self.nodeCount = 0 #counter to count nodes
    
    
    #Structure of array [parent, number of childs n, child data 1, child data 2.... child data n, count of child 1 children, data of children .... ]
    def createTree(self, arr): #Create a Tree
        if len(arr) == 0: #if arr is empty return None
            return None
        
        self.root = Node(arr[0]) #Root Node
        self.nodeCount +=1 #Increase count of nodes
        childCount = arr[1] #Number of childer following by data 
        q = [self.root] #push root node in queue
        i = 2 #Since 0th element is root and 1st is number of child so add up both
        
        while len(q)> 0: #while queue is not empty
            parent = q.pop(0) #getting parent Node
            for childData in range(childCount):
                if i> len(arr): #if Problem in input array child count
                    print("Some problem in child count please give input again")
                    break
                    
                child = Node(arr[i]) #making child as Node
                q.append(child) #add in queue
                i+=1 #increment array
                
                self.nodeCount += 1 #increment tree node counter
                parent.children.append(child) #add it in parent
                
            if i > len(arr)-1: #if no child and array is empty
                break
            childCount = arr[i] #Increment
            i += 1 #increase array index 
        return self.root #return root
    
    
    #print Tree 
    #parent : child1.data, child2.data ..... childN.data
    def printTree(self):        
        q = [self.root] #Add to queue        
        while len(q)> 0: #while queue is not empty
            parent = q.pop(0) #remove 
            print(parent.data, end =  " : ") #add parent data
            for child in parent.children:  #Iterate over children
                q.append(child) #add child too queue
                print(child.data, end = ", ") #print child
                
            print()
            
            
        
        
    def insertNode(self, parent, value): #insert when parent is given
        q = [self.root] #create a queue
        while len(q) > 0:#while q is not empty
            currParent = q.pop(0) #Get the parent
            if currParent.data == parent: #is parent matched with given data
                child = Node(value) #Get 
                currParent.children.append(child)
                print("Child", value, "add to parent", currParent.data)
                break #Break the loop
                
            for child in currParent.children: #add child to parent
                q.append(child)
                
        return self.root
            
        
        
    def sumNodeHelper(self, root): #Get the sum of all the nodes
        nodeSum = root.data #get the data
        for child in root.children:
            nodeSum +=  self.sumNodeHelper(child) #add sum to curr Sum and recursively call the tree
        return nodeSum #return the sum
            
    def sumNode(self):
        return self.sumNodeHelper(root) #Create a Helper function for easier recursive calls
    
    
    
    
    
    
    def largestNodeHelper(self, root): #get the largest node value
        maxValue = root.data #Assume root to be the max
        for child in root.children:
            maxValue = max(maxValue, self.largestNodeHelper(child)) #get the max and the child of max value
            
        return maxValue #return the final max value
    
    def largestNode(self):
        return self.largestNodeHelper(root)  #Create a Helper function for easier recursive calls
    
    
    
    
    
    
    def heightTreeHelper(self, root):
        currHeight = 1 #Height of root
        for child in root.children: #Call the child
            currHeight = max(currHeight, self.heightTreeHelper(child)+1) #Get the depth and increase the height of tree
        return currHeight #return curr Height
    
    def heightTree(self):
        return self.heightTreeHelper(root) #Create a Helper function for easier recursive calls
    
    
    
    
    
    
    def findNodeHelper(self, root, data):
        if root.data == data: #if root is data
            return True #return
            
        for child in root.children: #call the childrent
            if self.findNodeHelper(child, data): #find in childs
                return True
            
        return False #Nothing found
            
    def findNode(self, data):
        return self.findNodeHelper(root, data) #Create a Helper function for easier recursive calls

    
    def leafsCount(self, root):
        count = 0 #get the count
        if len(root.children) == 0: #if no child it means a leaf node 
            return 1 #return count 1
        
        for child in root.children: #Call the childs
            count+= self.leafsCount(child) #Return the count of all the leaf
        return count #return final count
    
    def leafs(self):
        return self.leafsCount(root) #Create a Helper function for easier recursive calls
    
    
    
    def maxSumHelper(self, root): #Get the max sum of child
        maxSum = sum(i.data for i in root.children) #Get the sum of childs
        for child in root.children: #call the childrent
            maxSum = max(maxSum, self.maxSumHelper(child)) #Get the max of currSum and sum of childs
    
        return maxSum #return max Sum
    
    def maxSum(self):
        return self.maxSumHelper(root)#Create a Helper function for easier recursive calls
```
