### Min Steps To 1

Given a positive integer 'n', find and return the minimum number of steps that 'n' has to take to get reduced to 1. You can perform any one of the following 3 steps:
1. Subtract 1 from it. (n = n - ­1) ,
2. If n is divisible by 2, divide by 2.( if n % 2 == 0, then n = n / 2 ) ,
3. If n is divisible by 3, divide by 3. (if n % 3 == 0, then n = n / 3 ).  

````python
sys.setrecursionlimit(10**6)
#Using recursion
def minStepRec(n):    
    if n == 1:
        return 0
    
    ans1 = float("inf")
    if n%3 == 0:
        ans1 = minStepRec(n//3)
        
    ans2 = float("inf")
    if n%2 == 0:
        ans2 = minStepRec(n//2)

    return 1 + min(ans1, ans2, minStepRec(n-1))



#Using memorization in recursion
def minStepRecDP(n, dp = {}):
    if n == 1:
        dp[0] = 0
        dp[1] = 0
        return 0
    
    ans1 = float('inf')
    if n - 1 not in dp:
        dp[n-1] = minStepRecDP(n-1, dp)
        ans1 = dp[n-1]
    else:
        ans1 = dp[n-1]
    
    
    
    ans2 = float("inf")
    if n%2 == 0:
        if n//2 not in dp:
            dp[n//2] = minStepRecDP(n//2, dp)
            ans2 = dp[n//2]
        else:
            ans2 = dp[n//2]

    
    
    
    ans3 = float("inf")
    if n%3 == 0:
        if n//3 not in dp:
            dp[n//3] = minStepRecDP(n//3, dp)
            ans3 = dp[n//3]
        else:
            ans3 = dp[n//3]
    
    return 1 + min(ans1, ans2, ans3)



#Using memorization in iteration
def minStepIter(n):
    memo = [float("inf")] * (n+1)
    memo[1] = 1
    memo[2] = 1
    memo[3] = 1
    i = 4
    while i < n+1:   
        ans1 = float('inf')
        ans2 = float('inf')
        ans3 = float('inf')
        
        ans1 = memo[i-1]
        if i%2 == 0:
            ans2 = memo[i//2]
        if i%3 == 0:
            ans3 = memo[i//3]
            
        memo[i] = 1 + min(ans1, ans2, ans3)
        i+=1
    return memo[n]



s = time.time()
n = 300
print("Result of recursive code for 100 inputs is :", minStepRec(100), "and time take to solve problem :", round((time.time() - s)*1000, 2))
s = time.time()
print("Result of memorization recursive code for 2000 inputs is :", minStepRecDP(2000), "and time take to solve problem :", round((time.time() - s)*1000, 5))
s = time.time()
print("Result of memorzation Iterative code for 2000 inputs is:", minStepIter(2000), "and time take to solve problem :", round((time.time() - s)*1000, 5))
````

### Minimum Number Of Squares

A number can always be represented as a sum of squares of other numbers. Note that 1 is a square and we can always break a number as  ```[(1 * 1) + (1 * 1) + (1 * 1) + …]```

Given a number n, find the minimum number of squares that sum to n.

````python
def minNoSqr(n):
    if n == 1 or n == 0:
        return 0
    root = int(n**.5)
    ans = sys.maxsize
    for i in range(1, root + 1):
        currAns = 1 + minNoSqr(n - i**2)
        ans = min(ans, currAns)
    
    return ans




def minNoSqrDp(n, dp = {}):
    if  n == 0:
        dp[0] = 0
        dp[1] = 1
        return 0
    
    
    root = int(n**.5)
    ans =  sys.maxsize
    for i in range(1, root+1):
        newCheckVal = n - i*i
        
        if newCheckVal not in dp:
            smallAns = minNoSqrDp(newCheckVal)
            dp[newCheckVal] = smallAns
            currAns = 1 + newCheckVal
        else:
            currAns = 1 + dp[newCheckVal]
            
        ans = min(ans, currAns)
        
    return ans





def minNoSqrIter(n):
    memo = {0: 0}
    for i in range(1, n + 1):
        memo[i] = sys.maxsize
        root = int(i**.5)
        for j in range(1, root+1):
            memo[i] = min(memo[i], 1+ memo[i - j*j])
    return memo[n]



s = time.time()
n = 300
print("Result of recursive code for 20 inputs is :", minNoSqr(20), "and time take to solve problem :", round((time.time() - s)*1000, 2))
s = time.time()
print("Result of memorization recursive code for 191 inputs is :", minNoSqrDp(500), "and time take to solve problem :", round((time.time() - s)*1000, 5))
s = time.time()
print("Result of memorzation Iterative code for 1000 inputs is:", minNoSqrIter(500), "and time take to solve problem :", round((time.time() - s)*1000, 5))
````

### LCS - Problem
<br>
<font face="Courier">Given two strings, 'S' and 'T' with lengths 'M' and 'N', find the length of the 'Longest Common Subsequence'.
For a string 'str'(per se) of length K, the subsequences are the strings containing characters in the same relative order as they are present in 'str,' but not necessarily contiguous. Subsequences contain all the strings of length varying from 0 to K.</font>

````python
def lisRec(str1, str2, m = 0, n = 0):
    if m == len(str1) or n == len(str2):
        return 0
    
    if str1[m] == str2[n]:
        ans =  1 + lisRec(str1, str2, m+1, n+1)
        
    else:
        ans1 = lisRec(str1, str2, m+1, n)
        ans2 = lisRec(str1, str2, m, n+1)
        ans = max(ans1, ans2)
        
    return ans



def lisRecDp(dp, str1, str2, m = 0, n = 0):
    if m == len(str1) or n == len(str2):
        return 0
    
    if str1[m] == str2[n]:
        if dp[m+1][n+1] == -1:
            smallAns = lisRecDp(dp, str1, str2, m+1, n+1)
            dp[m+1][n+1] = smallAns
            ans = 1 + smallAns
        else:
            ans = 1 + dp[m+1][n+1]
        
    else:
        if dp[m+1][n] == -1:
            ans1 = lisRecDp(dp, str1, str2, m+1, n)
            dp[m+1][n] = ans1
        else:
            ans1 = dp[m+1][n]
            
            
        if dp[m][n+1] == -1:
            ans2 = lisRecDp(dp, str1, str2, m, n+1)
            dp[m][n+1] = ans2
        else:
            ans2 = dp[m][n+1]
        ans = max(ans1, ans2)
        
    return ans





def lisMemo(str1, str2):
    m = len(str1)
    n = len(str2)
    memo = [[0]* (n+1)] * (m+1)
    
    for i in range(m):
        for j in range(n):
            if str1[i] == str2[j]:
                memo[i+1][j+1] = 1 + memo[i][j]
            else:
                memo[i+1][j+1] = max(memo[i+1][j], memo[i][j+1])
    return memo[m][n]




s1 = "abcdefgh"
s2 = "bdcijklabc"
s = time.time()
n = 300
print("Result of recursive code for s1 and s2 inputs is :", lisRec(s1, s2), "and time take to solve problem :", round((time.time() - s)*1000, 2))
s = time.time()
s1 = "qbpapenoqvsybadinwcboagufovosfs"
s2 = "bbxiwmbvfileownsayviebvodbapbda"
m = len(s1)
n = len(s2)
dp = [[-1]*(m+1)]*(n+1)
print("Result of memorization recursive code for s1 and s2 inputs is :", lisRecDp(dp, s1, s2), "and time take to solve problem :", round((time.time() - s)*1000, 5))
s = time.time()
print("Result of memorzation Iterative code for s1 and s2 inputs is:", lisMemo(s1, s2), "and time take to solve problem :", round((time.time() - s)*1000, 5))
````
